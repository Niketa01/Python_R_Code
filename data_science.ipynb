```python
#impute with mean
mhm['Price'].fillna(mhm['Price'].mean(),inplace=True)
mhm['Price'].describe()
mhm['Price'].value_counts()
mhm['Bedroom2'].fillna(mhm['Bedroom2'].mean,inplace=True)
pd.isna(mhm['Bedroom2']).sum()
#similarly, to do for other columns
```

```python
#outlier treatment
import matplotlib.pyplot as mp
%matplotlib inline
mp.plot(mhm['Distance'])
```

```python
mhm['Distance'].describe()
```

```python
Q1=mhm['Distance'].quantile(.25)
Q1
```

```python
Q3=mhm['Distance'].quantile(.75)
Q3
```

```python
IQR=Q3-Q1
IQR
```

```python
Floor=Q1-1.5*IQR
Floor
```

```python
Cap=Q3+1.5*IQR
Cap
```

```python
mhm_main=mhm[(mhm['Distance']>=Floor)&(mhm['Distance']<=Cap)]
mhm_main.head()
```

```python
mp.plot(mhm_main['Distance'])
```
