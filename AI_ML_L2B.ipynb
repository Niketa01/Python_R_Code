```python
# ECLAT ALGORITHMN
from itertools import chain, combinations
kot=0
freqItems=dict()
support=dict()
def eclat(prefix, items, dict_id):
    while items:
        i,itids=items.pop()
        isupp=len(itids)
        if isupp>=minsupp:
            freqItems[frozenset(prefix+[i])] = isupp
            suffix=[]
            for j, ojtids in items:
                jtids=itids&ojtids
                if len(jtids)>=minsupp:
                    suffix.append((j,jtids))
            dict_id+=1
            eclat(prefix+[i],sorted(suffix, key=lambda item: len(item[1])),dict_id)
```

```python
def rules(freqItems,confidence):
    Rules=[]
    cnt=0
    for items,support in freqItems.items():
        if len(items)>1:
            lst=list(items)
            antecedent=lst[:len(lst)-1]
            consequent=lst[-1:]
            conf=float(freqItems[frozenset(items)]/freqItems[frozenset(antecedent)])
            if conf>=confidence:
                cnt+=1
                Rules.append((antecedent,consequent,support,conf))
    print('found %d rules'%(cnt))
    return Rules        
```

```python
def print_frequent_itemsets(output_freqitems, freqItems):
    file=open(output_freqitems,'w+')
    for item,support in freqItems.items():
        file.write("{} : {}\n".format(list(item),round(support,4)))
    file.close()
```

```python
def print_rules(output_file,Rules):
    file=open(output_file,'w+')
    for ant,con,sup,conf in sorted(Rules):
        file.write("Rule: {} ==> {} : {} : {}\n".format((ant),(con),round(sup,4),round(conf,4)))
    file.close()
```

```python
def read_data(filename, delimiter=','):
    data={}
    trans=0
    f=open(filename,'r')
    for row in f:
        trans+=1
        for item in row.split(delimiter):
            if item not in data:
                data[item]=set()
            data[item].add(trans)
    f.close()
    return data
```

```python
if __name__=="__main__":
    minsupp=2
    confidence=30
    output_freqitems="output_freqitems.csv"
    output_file="output_rules.csv"
    dict_id=0
    data1=read_data(r"C:\Users\Niketa Mittal\Downloads\Market_Basket_Optimisation.csv",',')
    eclat([], sorted(data1.items(), key=lambda item: len(item[1]), reverse=True), dict_id)
    Rules=rules(freqItems,confidence)
    print('found %d frequent items'%len(freqItems))
    print('Writing rules.....')
    print_frequent_itemsets(output_freqitems,freqItems)
    print_rules(output_file,Rules)
    print('Files written successfully')
```

```python
#4 RECOMMENDATION SYSTEM USING MOVIE RATINGS DATASET
import math
import pandas as pd
import numpy as np
from scipy.sparse import csr_matrix
from sklearn.model_selection import KFold
from sklearn.neighbors import NearestNeighbors
import warnings
from fuzzywuzzy import fuzz
warnings.simplefilter("error")
df_ratings=pd.read_csv(r'C:\Users\Niketa Mittal\Downloads\ml-latest-small\ratings.csv')
df_movies=pd.read_csv(r'C:\Users\Niketa Mittal\Downloads\ml-latest-small\movies.csv')
df_tags=pd.read_csv(r'C:\Users\Niketa Mittal\Downloads\ml-latest-small\tags.csv')
```

```python
df_movie_count=pd.DataFrame(df_ratings.groupby('movieId').size(), columns=['count'])
popular_movies=list(df_movie_count[df_movie_count['count']>10].index)
movie_filter=df_ratings.movieId.isin(popular_movies).values
movie_filter
```

```python
df_user_count=pd.DataFrame(df_ratings.groupby('userId').size(), columns=['count'])
active_users=list(df_user_count[df_user_count['count']>10].index)
user_filter=df_ratings.userId.isin(active_users).values
user_filter
```

```python
df_rating_filtered=df_ratings[movie_filter&user_filter]
df_rating_filtered
```

```python
movie_user_matrix=df_rating_filtered.pivot(index='movieId',columns='userId',values='rating').fillna(0)
hashmap={movie:i for i,movie in
        enumerate(list(df_movies.set_index('movieId').loc[movie_user_matrix.index].title))}
movie_matrix_sparse=csr_matrix(movie_user_matrix.values)
movie_matrix_sparse
```

```python
def fuzzy_matching(hashmap, fav_movie):
    match_tuple=[]
    for title,indx in hashmap.items():
        ratio=fuzz.ratio(title.lower(),fav_movie.lower())
        if ratio>=60:
            match_tuple.append((title,indx,ratio))
    match_tuple=sorted(match_tuple,key=lambda x: x[2])[::-1]
    if not match_tuple:
        print("no match found")
    else:
        print('possible matches found: {0}\n'.format([x[0] for x in match_tuple]))
    return match_tuple[0][1]
```

```python
def inference(data,hashmap,fav_movie,n_recomm):
    model=NearestNeighbors(metric='cosine',algorithm='brute', n_neighbors=20, n_jobs=-1)
    model.fit(data)
    print('input movie is: ',fav_movie)
    idx=fuzzy_matching(hashmap,fav_movie)
    print('starting inference by recommendation system:')
    print('-------\n')
    #t0=time.time()
    distances,indices=model.kneighbors(data[idx], n_neighbors=n_recomm+1)
    raw_recomm=sorted(list(zip(indices.squeeze().tolist(),distances.squeeze().tolist())),key=lambda x: x[1])[:0:-1]
    return raw_recomm
```

```python
fav_movie='Iron Man'
top_n=10
recomm=inference(movie_matrix_sparse, hashmap,fav_movie,top_n)
rev_hashmap={v:k for k,v in hashmap.items()}
print('recommendations for {}: '.format(fav_movie))
for i, (idx,dist) in enumerate(recomm):
    print('{0}: {1}, with distance of {2}'.format(i+1,rev_hashmap[idx], dist))
```

```python
#6 K-MEANS CLUSTERING ON ONLINE RETAIL CUSTOMER DATA
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
%matplotlib inline
df=pd.read_excel(r'C:\Users\Niketa Mittal\Downloads\Online_Retail.xlsx', encoding='ISO-8859-1')
df=df.drop(['StockCode','Country','Description'], axis=1)
df.head()
```

```python
df.isna().sum()
df.shape
```

```python
df.dropna(axis=0,subset=['CustomerID'],inplace=True)
df.shape
```

```python
df['Amount']=df.Quantity*df.UnitPrice
df.head()
```

```python
df1=df.groupby('CustomerID').Amount.sum()
df1=df1.reset_index()
df1.shape
```

```python
df1.head()
```

```python
df2=df.groupby('CustomerID').InvoiceNo.count()
df2=df2.reset_index()
df2.shape
```

```python
df2.head()
```

```python
df3=pd.merge(df1,df2, on='CustomerID',how='inner')
df3.head()
```

```python
scX=StandardScaler()
df3=scX.fit_transform(df3)
df3.shape
```

```python
from scipy.spatial.distance import cdist
from sklearn.metrics import silhouette_score
distort=[]
r=range(2,10)
for i in r:
    Kcluster=KMeans(n_clusters=i).fit(df3)
    predicts=Kcluster.fit_predict(df3)
    sills=silhouette_score(df3,predicts,metric='euclidean')
    distort.append(sills)
plt.plot(r,distort,'bx-')
plt.xlabel('no. of clusters')
plt.ylabel('distortions')
plt.show()
```

```python
kmeans=KMeans(n_clusters=5)
y_pred=kmeans.fit_predict(df3)
# Visualising the clusters
plt.scatter(df3[y_pred == 0, 0], df3[y_pred == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(df3[y_pred == 1, 0], df3[y_pred == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(df3[y_pred == 2, 0], df3[y_pred == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(df3[y_pred == 3, 0], df3[y_pred == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(df3[y_pred == 4, 0], df3[y_pred == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of customers')
plt.legend()
plt.show()
```
